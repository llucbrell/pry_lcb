# Aplicando filtros para conocer la cuadrícula y las ondas
Los filtros se pueden aplicar para ejecutar multitud de acciones. Una de estas acciones es la detección de bordes. Es decir, cuando hay un cambio de intensidad importante. Esto es importante para la detección de la cuadrícula, así como también podemos aplicarlos para destacar aún más las ondas.

Es importante saber cuando aplicar estos filtros, pues a veces pueden aumentar el ruido de la imagen. Puedes consultar la documentación que habla de los filtros pulsando [aquí](https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_filtering/py_filtering.html "Ir a la Documentación").

### Creamos el código base para el programa

```python
import numpy as np
import matplotlib.image as mpimg  # librería usada para leer imágenes
import matplotlib.pyplot as plt
import cv2  # librería de visión por computador

%matplotlib inline

# Lee la imagen
impath ='imagenes/separa/e1.jpg'
# Cargamos la imagen en librería OpenCV
imagecv = cv2.imread(impath)
# Mostramos por pantalla las características de la imagen
print( 'La imagen es, ', type(imagecv), 'de dimensiones ', imagecv.shape )

imcopy = np.copy(imagecv)
imcopy = cv2.cvtColor(imcopy, cv2.COLOR_BGR2RGB)
imgris = cv2.cvtColor(imcopy, cv2.COLOR_RGB2GRAY)
plt.imshow(imcopy)
```

### Aplicamos los filtros
Se puede pasar un filtro que destaque sólo los cambios que se producen en horizontal. Este tipo de filtro nos ayudará a saber dónde está la referencia del voltaje en el electrocardiograma.

```python
# aplicando el filtro sobel

filtro_sY = np.array([ [-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

imgfiltrada = cv2.filter2D(imgris, -1, filtro_sY)
imgfiltY = np.copy(imgfiltrada)

# Muestra las imágenes
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.imshow(imcopy)
ax2.imshow(imgfiltrada, cmap='gray')
```

Podemos también hacer lo mismo para averiguar dónde está la cuadrícula en vertical. Es decir, con este método podemos destacar las líneas que hacen referencia al tiempo. Sobre estas líneas se podrían realizar múltiples cálculos, sacar el tamaño de los cuadrados en píxeles y tras esto, analizar sólamente la onda en función de estos.

```python
# aplicando el filtro sobel
filtro_sX = np.array([ [-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

imgfiltrada = cv2.filter2D(imgris, -1, filtro_sX)
imgfiltX = np.copy(imgfiltrada)

# Muestra las imágenes
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.imshow(imagecv)
ax2.imshow(imgfiltrada, cmap='gray')
```

También podemos aplicar un filtro que destaque tanto la vertical como la horizontal.

```python
# aplicando el filtro sobel
filtro_sXY = np.array([ [0, -1, 0],
                    [-1, 4, -1],
                    [0, -1, 0]])

imgfiltrada = cv2.filter2D(imgris, -1, filtro_sXY)
imgfiltXY = np.copy(imgfiltrada)
# Muestra las imágenes
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.imshow(imagecv)
ax2.imshow(imgfiltrada, cmap='gray')
```

### Conversión a imagen binaria
Podemos incluso destacar aún más la diferencia. Es decir que la imagen a formar sea sólo uno o cero, que exagere aún más los bordes, por ejemplo para encontrar la cuadrícula más grande, la formada por 5 cuadritos o unidades.

Puedes ver como usar la función [threshold](https://docs.opencv.org/2.4/modules/imgproc/doc/miscellaneous_transformations.html?highlight=threshold#threshold "Documentación") así como las [características](https://docs.opencv.org/2.4/doc/tutorials/imgproc/threshold/threshold.html "Operación") que tiene esta operación.

```python
# Puedes ver el resultado sobre imgfiltX, imgfiltY, imgfiltXY
retval, imgbin = cv2.threshold(imgfiltXY, 80, 255, cv2.THRESH_BINARY)
# Muestra las imágenes
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.imshow(imgfiltrada)
ax2.imshow(imgbin, cmap='gray')
```

###  Utilización de filtros en imágenes no formateadas
Se puede también utilizar filtros sobre imágenes no formateadas. Estos filtros nos pueden ayudar a crear mayores contrastes para obtener la información que más nos interese.

```python
# Lee la imagen
impath ='imagenes/separa/l3.jpg'
# Puedes usar la imagen l1.jpg , l2.jpg o l3.jpg
# Especialmente útil para imágenes con gran contraste como es el caso de l3.jpg con el filtro en X o en Y

# Cargamos la imagen en librería OpenCV
imagecv2 = cv2.imread(impath)
imcopy2 = np.copy(imagecv2)
imcopy2 = cv2.cvtColor(imcopy2, cv2.COLOR_BGR2RGB)
imgris2 = cv2.cvtColor(imcopy2, cv2.COLOR_RGB2GRAY)

# Aplicamos el filtro que queramos X, Y o XY
imgfiltrada2 = cv2.filter2D(imgris2, -1, filtro_sY)
imgfiltXY2 = np.copy(imgfiltrada2)
# Muestra las imágenes
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.imshow(imagecv2)
ax2.imshow(imgfiltrada2, cmap='gray')
```

Así, la aplicación de un filtro sobre imágenes en las que anteriormente no se obtenían buenos resultados, puede acontecer en una mejora sustancial de los mismos.
