# Usando el color

Si un elemento de una imagen es más fácil de reconocer por su color para nosotros, también lo debe ser para los ordenadores.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
%matplotlib inline
```

```python
# Leemos la imagen
image = mpimg.imread('imagenes/separa/e1.jpg')

plt.imshow(image)
```

```python
# Separamos cada canal y lo almacenamos en una variable
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

# Visualizamos los colores de manera separada
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.set_title('Canal R')
ax1.imshow(r, cmap='gray')
ax2.set_title('Canal G')
ax2.imshow(g, cmap='gray')
ax3.set_title('Canal B')
ax3.imshow(b, cmap='gray')

```

Podemos observar que al separar los colores, se acentúan y suavizan las líneas, dependiendo del canal.
