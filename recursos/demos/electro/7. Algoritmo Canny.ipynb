# Algoritmo de detección de bordes Canny
En este experimento, vamos a usar un algoritmo que incluye la libreía OpenCV para la detección de bordes. Este algoritmo, lo que hace es pasar a la imagen una serie de filtros para intentar identificar mejor los bordes.

Ahora se aplica a nuestras imágenes.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Lee la imagen
impath ='imagenes/separa/e1.jpg'
# Cargamos la imagen en librería OpenCV
imagecv = cv2.imread(impath)
# Mostramos por pantalla las características de la imagen
print( 'La imagen es, ', type(imagecv), 'de dimensiones ', imagecv.shape )

imcopy = np.copy(imagecv)
imcopy = cv2.cvtColor(imcopy, cv2.COLOR_BGR2RGB)
imgris = cv2.cvtColor(imcopy, cv2.COLOR_RGB2GRAY)
plt.imshow(imcopy)
```

### Aplicamos el algoritmo

```python
uno_a_tres = cv2.Canny(imgris, 10, 100)
uno_a_dos = cv2.Canny(imgris, 350, 500)
 
    
# Display the images
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('Algoritmo con mayor diferéncia de valores')
ax1.imshow(uno_a_tres, cmap='gray')

ax2.set_title('Algoritmo con menor diferencia entre valores')
ax2.imshow(uno_a_dos, cmap='gray')
```

Se aprecia un resultado bastante bueno en la detección de la cuadrícula en la imagen de la izquierda. En la de la derecha en cambio se aprecia una buena detección de la onda. 

## Conclusión
Este algoritmo es bastante útil para separar las ondas del fondo.
